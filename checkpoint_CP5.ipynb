2 – Qual seria o tamanho de amostra para uma pesquisa em todos os municípios do Brasil
contidos no dataframe? (Utilizar a variância da renda per capita como base da fórmula de
amostragem, z = 1,96 e margem de erro = 20).

```python
dados_amostra_br = dados_br.agg(media_renda = pd.NamedAgg('RDPC', 'mean'),
                                   dp_renda = pd.NamedAgg('RDPC', 'std'),
                                   n_municipios = pd.NamedAgg('RDPC', 'count'))
```

```python
dados_amostra_br
```

```python
def formula_amostra_continua(ME, N, Z, S):

  n = (Z**2 * S**2 * N) / ((ME**2 * (N-1)) + (Z**2 * S**2))

  return int(n)
```

```python
ME = 20 # margem de erro
Z = 1.96 # 95% de confiança
S = 243.269333
N = 5565 # total de municipios
```

```python
n_br = formula_amostra_continua(ME, N, Z, S)
n_br
```

```python
def formula_amostra_continua_sp(ME_sp, N_sp, Z_sp, S_sp):

  n = (Z_sp**2 * S_sp**2 * N_sp) / ((ME_sp**2 * (N_sp-1)) + (Z_sp**2 * S_sp**2))

  return int(n)
```

```python
formula_amostra_continua_sp(ME_sp = 100, N_sp = 15000, Z_sp = 1.96, S_sp = 500)
```

```python
ME_sp = 20 # margem de erro
Z_sp = 1.96 # 95% de confiança
S_sp = 197.398270
N_sp = 645 # total de municipios
```

```python
n_sp = formula_amostra_continua_sp(ME_sp, N_sp, Z_sp, S_sp)
n_sp
```
